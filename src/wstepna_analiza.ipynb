```python
import pandas as pd
```

# Wczytywanie pliku CSV

```python
df = pd.read_csv('National_Health_and_Aging_Trends_Study__NHATS____Vision_and_Eye_Health_Surveillance_System__VEHSS_.csv')
```

# Wyświetlenie tabeli

```python
print(df)
```

```python
# Sprawdzenie jakie są możliwe odpowiedzi 
unikalne = df["Response"].unique()

print(unikalne)
```

```python
# Sprawdzanie jakie są możliwe odp. dla VIS bliż
# filtr: tylko wiersze gdzie inna_kolumna == "x"
przefiltrowane1 = df[df["Question"] == "Presenting near acuity"]

# unikalne wartości z wybranej kolumny
unikalne1 = przefiltrowane1["Response"].unique()

print(unikalne1)
```

```python
# Sprawdzanie jakie są możliwe odp. dla VIS dal
# filtr: tylko wiersze gdzie inna_kolumna == "x"
przefiltrowane2 = df[df["Question"] == "Presenting visual acuity"]

# unikalne wartości z wybranej kolumny
unikalne2 = przefiltrowane2["Response"].unique()

print(unikalne2)
```

```python
# Sprawdzanie jakie są możliwe odp. dla kontrastu
# filtr: tylko wiersze gdzie inna_kolumna == "x"
przefiltrowane3 = df[df["Question"] == "Contrast sensitivity"]

# unikalne wartości z wybranej kolumny
unikalne3 = przefiltrowane3["Response"].unique()

print(unikalne3)
```

```python
# dopasowanie odpoweidzi do id_odp
wynik = df.groupby("Response")["ResponseID"].unique()

print(wynik)
```

```python
# zakresy grup wiekowych
unikalne4 = df["Age"].unique()

print(unikalne4)
```

```python
# podział płci
unikalne5 = df["Sex"].unique()

print(unikalne5)
```

```python
# podział na rasy
unikalne6 = df["RaceEthnicity"].unique()

print(unikalne6)
```

```python
# ile jest czynników zwiększających ryzyko
unikalne7 = df["RiskFactor"].unique()

print(unikalne7)
```
